#### 1. Podatki in opis problema
V dokumentu so predstavljene razne ugotovitve in zanimivi razultati, ki so rezultat obdelave podatkov evidence o nesrečah na smučiščih v Sloveniji.

---

Podatki, ki so bili uporabljeni za obdelavo, so bili pridobljeni iz OPSI ([Odprti podatki Slovenije](https://podatki.gov.si/)), in sicer iz [Evidence nesreč na smučiščih][1].

Podatki so zapisani v dveh datotekah - ena beleži nesreče v obdobju med 1. 12. 2013 in 9. 3. 2017, druga pa v obdobju med 1. 11. 2017 in  2. 4. 2019. V datotekah so podatki zapisani v različnih formatih - nekatere statistike lahko izluščimo iz obeh tabel, nekateri atributi pa so prisotni le v eni izmed tabel. Zaradi tega se bo za nekatere ugotovitve uporabljala le ena izmed obeh tabel.

V tabeli je shranjenih veliko različnih atributov, kar nam bo omogočalo ugotoviti zelo zanimive povezave med podatki. Prva tabela ima poleg osnovnih podatkov (datum in čas, resnost poškodbe) predvsem tudi podatke o beleženju incidenta (skica, fotografija, prisotnost zdravnika, prisotnost nadzornika). Druga tabela pa se osredotoča bolj na podrobno beleženje nastalih poškodb ter na specifično smučišče, kjer je prišlo do nesreče.

Podatki so bili obdelani (in rezultati predstavljeni) v [notebooku](PR01.ipynb), dodatni grafi in tabele pa so zapisani [v prilogi](supplement.ipynb).

#### 3. Starost in spol ponesrečencev
Kako so smučarji in deskarji različnih starostnih skupin zastopani v statistiki nesreč?

![alt text](./res/age_distribution_normalized.png "Porazdelitve zabeleženih nesreč glede na starost in zvrst")

Na zgornjem [grafu](./supplement.ipynb#Graf-distribucije-starosti-ponesre%C4%8Dencev-(normaliziran):) lahko razberemo posamezen delež zastopanosti starostne skupine (in podstkupine zvrsti) v statistiki nesreč na smučiščih. Vidimo, da je največ ponesrečencev v starostni skupini od 10 do 19 let. Poleg tega je največ ponesrečencev smučarjev, pri mladostnikih pa nezanemarljiv delež poškodovancev predstavljajo tudi deskarji.

Podatke lahko tudi pogledamo na [nenormaliziranem grafu](./supplement.ipynb#Graf-distribucije-starosti-ponesre%C4%8Dencev:).

Na voljo je tudi [graf](./supplement.ipynb#Graf-distribucije-spola-ponesre%C4%8Dencev-(normaliziran):), ki ponesrečence razdeli glede na spol. Iz tega grafa lahko razberemo, da sta spola skoraj enakovredno predstavljena v statistiki - skupaj je 959 moških in 933 žensk. Razvidno je tudi, da je delež deskarjev, ki so se ponesrečili manjši med ženskami kot pa med moškimi. Tu je vprašanje, a to pomeni, da se ženske, ki deskajo, bolj pogosto poškodujejo (oz. ženske, ki smučajo, manj pogosto), ali pa je le žensk, ki deskajo, manj.

#### 4. Časovna razčlenitev
Oglejmo si pogostost nesreč glede na različne časovne razčlenitve.

![alt text](res/month_year_distribution.png "Nesreče glede na mesec in leto")

Na [grafu](./supplement.ipynb#Graf-distribucije-nesre%C4%8D-glede-na-mesec-in-leto:) si lahko pogledamo, kako se število nesreč spreminja skozi leta. Opazimo lahko nekaj posebnih primerov: Februar 2015 in Februar 2019 sta imela bistveno več nesreč kot pa druga leta v istem obdobju, leta 2014 pa sta bila Januar in December dokaj bolj varna v primerjavi z drugimi leti.

Če podatke razdelimo glede na [dan v letu](./supplement.ipynb#Graf-distribucije-nesre%C4%8D-glede-na-dan-v-letu:), si lahko bolj podrobneje pogledamo časovne trende skozi leto. Na grafu vidimo razne vrhove in doline, ki mogoče predstavljajo več kot le naključen šum. Ti podatki v veliki meri korelirajo s sezono smučanja (v toplejših mesecih se ne smuča). Zanimivo pa je izpostaviti tudi to, da je bilo nekaj nesreč zabeleženih tudi poleti.

Podatke lahko razdelimo tudi glede na [dan v tednu](./supplement.ipynb#Graf-distribucije-nesre%C4%8D-glede-na-dan-v-tednu:). Vidimo, da je najmanj nesreč zabeleženih na ponedeljke. Čez teden število nesreč raste, v soboto pa jih je največ. Sobota je skoraj dvakrat bolj nevarna (kar se tiče nesreč na smučiščih) kot pa ponedeljek - približno 19% proti približno 10%.

Na koncu si oglejmo še porazdelitev nesreč glede na [uro v dnevu](./supplement.ipynb#Graf-distribucije-nesre%C4%8D-glede-na-uro-v-dnevu:). V nočnih urah nesreč skoraj ni - smučišča so takrat verjetno zaprta, a si še vseeno kdo rad privošči nočno smučanje (in se poškoduje). Pogostost poškodb začne naraščati od približno devete ure dopoldne in doseže vrh ob približno enajstih. Nato začne pogostost nesreč spet padati do četrte ure popoldne. Od štirih pa do nekje enajstih popoldne pa nastane še en 'hrib' nesreč - ker se smuča v zimskih mesecih in ker je pozimi dan zelo kratek, je verjetno razlog za to tema.

#### 5. Državnost ponesrečencev
Iz katerih držav so udeleženci nesreč na smučiščih?

![alt text](res/country_distribution.png "Državnost ponesrečencev")

Iz [grafa](./supplement.ipynb#Graf-dr%C5%BEavnosti-ponesre%C4%8Dencev:) je razvidno, da je velika večina ponesrečencev Slovencev (kot bi pričakovali). Na drugem mestu je Hrvaška - kljub temu, da so Slovenska smučišča bližja Italiji in Avstriji. Razlog za to verjetno tiči v tem, da imajo Italijani in Avstrijci v svojih državah veliko smučišč, na Hrvaškem pa jih skoraj ni.

Zanimivo je tudi relativno veliko število poškodovancev iz Anglije. Zanimivo bi bilo izvedeti, ali je to zaradi tega, ker Angleži pogosto hodijo smučat v Slovenijo, ali pa so le bolj dovzetni za poškodbe.

Italija in Avstrija sta kljub geografski bližini našim smučiščem predstavljeni v relativno manjšem številu - verjetno njihovi državljani smučajo doma, mogoče pa so zelo dobri pri izogibanju nesreč na smučiščih.

Prvih pet stolpcev (dan v tednu, čas, zvrst, vreme, ravnanje) predstavlja atribute, katere bomo uporabili v odločitvenem drevesu za predvidevanje atributa v zadnjem stolpcu (resnost poškodbe).

Pri ustvarjanju odločitvenega drevesa se omejimo na le nekaj listov, saj hočemo, da bo rezultat pregleden in se bomo iz tega lahko kaj naučili.

![alt text](res/decision_tree.png "Odločitveno drevo za resnost poškodbe")

Zgoraj vidimo [odločitveno drevo](./supplement.ipynb#Graf-odlo%C4%8Ditvenega-drevesa-za-resnost-po%C5%A1kodbe:), ki poskuša ugotoviti resnost poškodbe glede na določene atribute. Bolj uporabno bi bilo sicer drevo, ki nam pove, katere okoliščine najbolj pripomorejo k nastanku poškodb, a vendar takih podatkov nimamo (seveda ne beležimo ne-nesreč). To odločitveno drevo pa nam še vseeno zadostuje - če se že poškodujemo, je bolje, da je poškodba lažja kot pa hujša. Lahko tudi sklepamo, da situacije, ki povzročajo večji delež lažjih poškodb, povzročajo tudi manj poškodb v splošnem. 

Zaradi tega, ker je velik delež vseh poškodb lažjih (94.4%), je težko sestaviti drevo, ki bi za določene atribute predvidel, da bo prišlo do hujše poškodbe - a nam vseeno da uporabne podatke, saj so situacije, kjer je delež hujših poškodb manjši, bolj varne.

Iz korena grafa lahko razberemo, da pri padcih pride do hujših poškodb v večji meri kot pa pri drugih načinih obnašanja med nesrečo. Iz obeh otrok lahko razberemo, da je jasno oz. sončno vreme bolj varno od oblačnega. Iz njunih otrok pa lahko glede na čas dneva ugotovimo, da prihaja do hujših poškodb bolj pogosto dopoldne.
